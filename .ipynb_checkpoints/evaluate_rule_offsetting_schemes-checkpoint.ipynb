```python
# Evaluating Rule-Offsetting Schemes for Sustainable Policy Growth in Modern Democracies
# Developed using Python by Heider Jeffer

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Step 1: Manually Creating Data
# Sample data (Country, Year, Offsetting Scheme Effectiveness, Policy Growth)
data = {
    'Country': ['Country1', 'Country2', 'Country3', 'Country4', 'Country5', 'Country6', 'Country7', 'Country8', 'Country9', 'Country10'],
    'Year': [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019],
    'OffsettingSchemeEffectiveness': [3.5, 4.2, 3.8, 4.5, 4.0, 4.1, 3.9, 4.3, 4.0, 4.4],
    'PolicyGrowth': [5.5, 6.0, 5.7, 6.3, 6.1, 6.4, 6.2, 6.5, 6.3, 6.8]
}

# Convert the dictionary to a pandas DataFrame
data_df = pd.DataFrame(data)

# Step 2: Data Cleaning (if necessary)
# For the sake of simplicity, let's assume the data is clean and doesn't need major cleaning steps.

# Step 3: Feature Selection/Engineering
# In this case, we assume 'PolicyGrowth' is the dependent variable (target),
# and other columns are features, including 'OffsettingSchemeEffectiveness' and 'Year'.
X = data_df[['OffsettingSchemeEffectiveness', 'Year']]  # Features
y = data_df['PolicyGrowth']  # Target variable

# Step 4: Feature Scaling
# Standardize the features to have zero mean and unit variance (important for many ML algorithms)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 5: Model Training
# Split the data into training and testing sets (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Create a linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Step 6: Model Evaluation
# Predict on the test set
y_pred = model.predict(X_test)

# Evaluate the model's performance
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"R-squared (RÂ²): {r2}")

# Step 7: Visualize Results with Multiple Plots

# Plot 1: Scatter Plot of True vs Predicted Policy Growth
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, color='blue')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--')  # Line of perfect prediction
plt.xlabel('True Values (Policy Growth)')
plt.ylabel('Predicted Values (Policy Growth)')
plt.title('True vs Predicted Policy Growth. Developed using Python by Heider Jeffer')
plt.grid(True)
plt.show()

# Plot 2: Distribution of Policy Growth
plt.figure(figsize=(10, 6))
plt.hist(data_df['PolicyGrowth'], bins=10, color='green', edgecolor='black')
plt.xlabel('Policy Growth')
plt.ylabel('Frequency')
plt.title('Distribution of Policy Growth. Developed using Python by Heider Jeffer')
plt.grid(True)
plt.show()

# Plot 3: Effectiveness vs Policy Growth (Scatter Plot)
plt.figure(figsize=(10, 6))
plt.scatter(data_df['OffsettingSchemeEffectiveness'], data_df['PolicyGrowth'], color='purple')
plt.xlabel('Offsetting Scheme Effectiveness')
plt.ylabel('Policy Growth')
plt.title('Offsetting Scheme Effectiveness vs Policy Growth. Developed using Python by Heider Jeffer')
plt.grid(True)
plt.show()

# Plot 4: Regression Line for Effectiveness vs Policy Growth
plt.figure(figsize=(10, 6))
plt.scatter(data_df['OffsettingSchemeEffectiveness'], data_df['PolicyGrowth'], color='purple')
# Plot the regression line
line_x = np.linspace(min(data_df['OffsettingSchemeEffectiveness']), max(data_df['OffsettingSchemeEffectiveness']), 100)
line_y = model.predict(scaler.transform(np.column_stack((line_x, [2010] * 100))))  # Using year 2010 for simplicity
plt.plot(line_x, line_y, color='red', linestyle='--')
plt.xlabel('Offsetting Scheme Effectiveness')
plt.ylabel('Policy Growth')
plt.title('Regression Line for Offsetting Scheme Effectiveness vs Policy Growth. Developed using Python by Heider Jeffer')
plt.grid(True)
plt.show()

# Plot 5: Predicted vs True Policy Growth
plt.figure(figsize=(10, 6))
plt.bar(range(len(y_test)), y_test - y_pred, color='orange')
plt.xlabel('Test Data Points')
plt.ylabel('Error (True - Predicted)')
plt.title('Prediction Error (True - Predicted Policy Growth). Developed using Python by Heider Jeffer')
plt.grid(True)
plt.show()

# Plot 6: Residuals Plot
residuals = y_test - y_pred
plt.figure(figsize=(10, 6))
plt.scatter(y_pred, residuals, color='blue')
plt.axhline(0, color='red', linestyle='--')  # Line of zero residual
plt.xlabel('Predicted Policy Growth')
plt.ylabel('Residuals')
plt.title('Residuals Plot. Developed using Python by Heider Jeffer')
plt.grid(True)
plt.show()

# Save the model for future use (optional)
import joblib
joblib.dump(model, 'policy_growth_model.pkl')

# Additional steps can be added, such as saving the processed data
# data_df.to_csv('processed_data.csv', index=False)
```

Here is a simple step-by-step explanation of the code:

### 1. **Import Libraries**:
   The first section imports the required libraries for handling data, performing machine learning tasks, and generating plots:
   - **pandas**: Used to handle and manipulate tabular data.
   - **numpy**: Used for numerical computations and working with arrays.
   - **matplotlib**: Used for generating plots and visualizations.
   - **sklearn**: A popular machine learning library providing algorithms, data preprocessing tools, and metrics.

### 2. **Manually Creating Data**:
   - A dictionary `data` is created with sample data for 10 countries. It includes information on:
     - **Country**: The name of the country.
     - **Year**: The year corresponding to the data point.
     - **OffsettingSchemeEffectiveness**: A numerical value representing how effective a policy is.
     - **PolicyGrowth**: The growth of policy, which is our target variable.
   - This dictionary is then converted into a pandas DataFrame (`data_df`), which makes it easier to work with.

### 3. **Feature Selection/Engineering**:
   - Here, we select the features (independent variables) and the target (dependent variable):
     - `X` includes **OffsettingSchemeEffectiveness** and **Year**, which will be used to predict the target variable.
     - `y` is the target variable, **PolicyGrowth**, which we want to predict based on the features.

### 4. **Feature Scaling**:
   - Feature scaling is applied using **StandardScaler** from `sklearn`. This standardizes the features (i.e., scales them to have zero mean and unit variance) to ensure they have the same scale for machine learning models.
   - After scaling, the scaled data is stored back as a DataFrame (`X_scaled`) to preserve the column names.

### 5. **Model Training**:
   - **Train/Test Split**: The data is split into two parts:
     - **80% training data** and **20% testing data**.
   - **Linear Regression Model**: A linear regression model is created and trained using the training data (`X_train` and `y_train`).
   - The model learns the relationship between **OffsettingSchemeEffectiveness** and **PolicyGrowth**.

### 6. **Model Evaluation**:
   - The model makes predictions on the test data (`X_test`).
   - **Performance Metrics** are calculated:
     - **Mean Squared Error (MSE)**: Measures the average squared difference between predicted and actual values.
     - **Root Mean Squared Error (RMSE)**: The square root of MSE, gives the error in the same unit as the target variable.
     - **R-squared (RÂ²)**: A measure of how well the model explains the variance in the target variable. A value closer to 1 indicates a better fit.

### 7. **Visualizing Results**:
   - Several plots are generated to visualize the model's performance and the data:
   
   - **Plot 1: True vs Predicted Policy Growth**: A scatter plot that compares the true values (actual policy growth) with the predicted values (from the model). A red dashed line represents a perfect prediction (where true values equal predicted values).
   
   - **Plot 2: Distribution of Policy Growth**: A histogram showing the frequency distribution of the target variable (`PolicyGrowth`), helping to understand its distribution.
   
   - **Plot 3: Effectiveness vs Policy Growth**: A scatter plot that shows the relationship between **OffsettingSchemeEffectiveness** and **PolicyGrowth**.
   
   - **Plot 4: Regression Line for Effectiveness vs Policy Growth**: A scatter plot of the same variables as in Plot 3, but this time a red regression line is added to visualize the model's predicted relationship.
   
   - **Plot 5: Predicted vs True Policy Growth**: A bar plot that shows the error (difference between true and predicted values) for each test data point.
   
   - **Plot 6: Residuals Plot**: A scatter plot that shows the residuals (the difference between predicted and actual values) against the predicted values. A horizontal red line at zero shows the expected residual for perfect predictions.

### 8. **Saving the Model**:
   - The trained model is saved using **joblib**. This allows the model to be loaded and used later for making predictions on new data.
